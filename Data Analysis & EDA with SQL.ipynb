```python
print("Number Of Vehicles ON Road According to Brand ")
%sql SELECT BRAND ,COUNT(BIKE_NAME) FROM BIKE GROUP BY BRAND ORDER BY COUNT(BIKE_NAME) DESC;
```

```python
%sql Select BRAND,AVG(Price) from Bike GROUP BY BRAND ORDER BY AVG(PRICE) ;
```

```python
%sql ( ( SELECT * FROM BIKE WHERE PRICE BETWEEN 40000 AND 70000 AND POWER>140 limit 10 ) ORDER BY POWER DESC ) ORDER BY AGE;
```
